```python
from sklearn.model_selection import train_test_split
df
```

```python
X = df.drop(["target","flower name"],axis ="columns")
X
```

```python
y = df.target
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
df.head()
```

```python
#our prediction 

model.predict([[5.0,3.6,1.4,0.2]])
```

```python
#tuning the parameters 
# c = regularization 
model_C = SVC(C = 10)
model_C.fit(X_train,y_train)
model_C.score(X_train,y_train)
```

```python
#tuning the parameters 
# c = regularization 
model_G = SVC(gamma = 6)
model_G.fit(X_train,y_train)
model_G.score(X_train,y_train)

```

```python
#model_linear_kernel = SVC(kernel= "poly")
#model_linear_kernel = SVC(kernel= "rbf")
model_linear_kernel = SVC(kernel= "linear")
#model_linear_kernel = SVC(kernel= "sigmoid")
model_linear_kernel.fit(X_train,y_train)
model_linear_kernel.score(X_train,y_train)
```
