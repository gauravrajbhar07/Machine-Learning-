# what is gradient_descent?




Is an algorithm that finds best fit line for given training data set

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y):
    m_curr = b_curr = 0; #random values
    iterations = 1000
    n = len(x)
    learning_rate = 0.01    
    plt.scatter(x,y,color='red',marker= '+',linewidth = 5)
    
    for i in range(iterations):
        yp = m_curr * x + b_curr
        
        plt.plot(x,yp,color="green")
        cost = (1/n) * sum([val**2 for val in (y - yp)])   #mse
        md = -(2/n) * sum(x*(y-yp))
        bd = -(2/n) * sum(y-yp)
        m_curr = m_curr - learning_rate* md
        b_curr = b_curr - learning_rate * bd
        
        print("m {} ,b {} ,cost {} ,iterations {} ".format(m_curr,b_curr,cost,i))
        
        #for stoping the iteration we can compare the 2 consicutive itteration 
    
    pass
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

#both the values are dependent of each other by a function 
#that can be obtain by the finding the best fit line 
```

```python
gradient_descent(x,y)
```
